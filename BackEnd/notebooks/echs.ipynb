```python
import cv2
from PIL import Image
```

```python
img = cv2.imread("../myMedic/backend/Mymedic/echs.jpg", flags=cv2.IMREAD_GRAYSCALE)
```

```python
_, new_img = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
Image.fromarray(new_img).show()  #popup view, doesnot work in colab
```

```python
import matplotlib.pyplot as plt
plt.imshow(new_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#adaptive made it better

new_img = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    61,
    11
)
```

```python
plt.imshow(new_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import pytesseract
```

```python
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
text = pytesseract.image_to_string(new_img, lang='eng')
print(text)
```
